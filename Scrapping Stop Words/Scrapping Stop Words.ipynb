### Scrapping Stop Words In English


```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException


url = "https://www.ranks.nl/stopwords"
driver = webdriver.Chrome(executable_path=r"C:\Users\shrut\Downloads\chromedriver.exe")
driver.get(url)

soup = BeautifulSoup(driver.page_source,'html.parser')
print(soup)
```

```python
def get_defualt_english_stop_words(soup):
        
        '''
        Returns the default english Stop Words
        
        '''
        a=soup.findAll("table")[0]
        first_td  = a.find('td')
        second_td = a.findAll('td')[1]


        def_stop_words=[]
        for dstop_word in first_td:

            if str(dstop_word)=="<br/>":
                continue
            def_stop_words.append(dstop_word.strip())

        for dstop_word in second_td:

            if str(dstop_word)=="<br/>":
                continue
            def_stop_words.append(dstop_word.strip())

        return def_stop_words
    
def_stop_words=get_defualt_english_stop_words(soup)
print(len(def_stop_words))
def_stop_words[:10]
```

```python
def get_mysql_stop_words(soup):
            
    '''Getting Mysql Stop Words'''
    
    b=soup.findAll("table")[1]
    mysql_stop_words=[]
    for i in b.findAll("td"):
        mysql_stop_words.append(i.text.strip())
    return mysql_stop_words

mysql_stop_words=get_mysql_stop_words(soup)
print(len(mysql_stop_words))
mysql_stop_words[:10]
```

```python
def get_goog_stop_words(soup):
    
        ''' Google History Stop Words '''
        
        c=soup.findAll("table")[2]
        goog_stop_words=[]
        first_td = c.find_all('div')

        for dstop_word in first_td:

            goog_stop_words.extend(dstop_word.text.split())
            if str(dstop_word)=="<br/>":
                continue

        return goog_stop_words
    
goog_stop_words=get_goog_stop_words(soup)
print(len(goog_stop_words))
goog_stop_words[:10]
```

```python
def get_long_list(soup):
    
    
    ''' Returns Long List of Stop Words '''
    
    d=soup.findAll("table")[3]
    
    lfirst_td  = d.find('td')
    lsecond_td = d.findAll('td')[1]
    lthird_td = d.findAll('td')[2]

    list_stop_words=[]

    for dstop_word in lfirst_td:


        if str(dstop_word)=="<br/>":
            continue
        list_stop_words.append(dstop_word.strip())

    for dstop_word in lsecond_td:

        if str(dstop_word)=="<br/>":
            continue
        list_stop_words.append(dstop_word.strip())

    for dstop_word in lthird_td:

        if str(dstop_word)=="<br/>":
            continue
        list_stop_words.append(dstop_word.strip())

    return list_stop_words


list_stop_words=get_long_list(soup)
print(len(list_stop_words))
list_stop_words[:10]
```

```python
import pandas as pd
```

```python
stop_words=[]
stop_words.extend(def_stop_words)
stop_words.extend(mysql_stop_words)
stop_words.extend(goog_stop_words)
stop_words.extend(list_stop_words)

df=pd.Series(stop_words,name="Stop Words")
```

```python
# There are many words that were repeated, therefore we need to drop duplicates
df.value_counts()
```

```python
df.drop_duplicates(inplace=True)
```
